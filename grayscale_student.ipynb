# Aufgabe: Farbbild nach Graustufen konvertieren

In dieser Aufgabe konvertierst du ein JPG-Bild mithilfe der **Rec. 709-Formel** in ein Graustufenbild.

Die Formel gewichtet die drei RGB-Kanäle entsprechend der menschlichen Farbwahrnehmung:

$$Y = 0.2126 \cdot R \;+\; 0.7152 \cdot G \;+\; 0.0722 \cdot B$$

> 💡 **Hinweis:** Grün wird am stärksten gewichtet, weil das menschliche Auge auf Grün am empfindlichsten reagiert.

## Schritt 1 — Bibliotheken importieren und Bild einlesen

Wir nutzen `PIL` zum Einlesen und `NumPy` für die Matrixoperationen.
Das Bild hat nach dem Einlesen die Form `(H, W, 3)` — Höhe × Breite × 3 Farbkanäle.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img     = np.array(Image.open("bild.jpg"))         # Shape: (H, W, 3), dtype uint8
r, g, b = img[:, :, 0], img[:, :, 1], img[:, :, 2]
```

## Schritt 2 — Rec. 709 Konvertierung

Wende die Formel auf die drei Kanäle an. Das Ergebnis ist ein 2D-Array mit Werten zwischen 0 und 255.
Konvertiere es abschliessend zu `uint8`.

```python
gray = 0.2126 * r + 0.7152 * g + 0.0722 * b        ##BLANKgray = ... (Rec.-709-Formel: Y = 0.2126*R + 0.7152*G + 0.0722*B)##
gray = gray.astype(np.uint8)                       
```

## Schritt 3 — Anzeige

Zeige Original und Graustufen nebeneinander an.
Verwende `cmap="gray"` damit Matplotlib das 2D-Array als Graustufenbild darstellt.

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].imshow(img)
axes[0].set_title("Original (RGB)")

axes[1].imshow(gray, cmap="gray")
axes[1].set_title("Graustufen (Rec. 709)")

plt.tight_layout()
plt.show()
```
