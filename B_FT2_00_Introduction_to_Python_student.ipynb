## Wichtige Datentypen in der Bildverarbeitung

| Typ | Wertebereich | Einsatz |
|---|---|---|
| `uint8` | 0 ... 255 | Viel Bilder auf Festplatte (JPEG, PNG) |
| `uint16` | 0 ... 65535 | RAW-Sensordaten |
| `float64` | -inf ... +inf | Alle Berechnungen |

**Achtung:** `uint8` sättigt bei 0 und 255.
In NumPy muss man **`np.clip()` vorher aufrufen**, um das explizit zu machen.


````python
# Hier sind ein paar Beispielwerte:
e = np.array([-5, 0, 1.49, 1.5, 1.51, 3, 255, 300])

e_uint8 = e.astype(np.uint8) # Konvertieren nach uint8. 

w = 6
print('Original float :', ''.join(f'{v:{w}}' for v in e))
print('Als uint8      :', ''.join(f'{v:{w}}' for v in e_uint8.astype(float)))
print()
print('Was passiert mit Werten unter 0 und über 255? Warum?')
````

````python

e_uint8 = np.clip(e, 0, 255).astype(np.uint8)   # Daher immer auf 0...5 clippen — dann erst casten!

w = 6
print('Original float :', ''.join(f'{v:{w}}' for v in e))
print('Als uint8      :', ''.join(f'{v:{w}}' for v in e_uint8.astype(float)))
print()
print('Beobachtungen:')
print('  -5   ->  0    (Saettigung nach unten)')
print('  300  ->  255  (Saettigung nach oben)')
print('  0.51 ->  0    (Trunkierung, nicht Rundung!)')
````

## Gamut Plot

Das Bild wird in eine Liste von (R,G,B)-Punkten umgewandelt und im 3D-Farbraum geplottet.
So sieht man, welche Farben im Bild tatsächlich vorkommen.

```matlab
% MATLAB:
tmpImg = imresize(img, 0.5);
tmpImg = double(reshape(tmpImg, [], 3));
scatter3(tmpImg(:,1), tmpImg(:,2), tmpImg(:,3), 3, tmpImg/255)
```


````python
# Bild verkleinern damit nicht Millionen Punkte gezeichnet werden müssen
tmp_img = img[::2, ::2, :]                        # jede 2. Zeile und Spalte nehmen

# Bild in einen N×3-Vektor umformen
pts = tmp_img.reshape(-1, 3).astype(float)        # shape: (N, 3)

# RGB-Farben als Liste von Strings
colors = [f'rgb({int(r)},{int(g)},{int(b)})' for r, g, b in pts]

fig = px.scatter_3d(
    x=pts[:,0], y=pts[:,1], z=pts[:,2],
    title='Gamut Plot — mit der Maus drehen'
)

fig.update_traces(marker=dict(
    size=1,
    opacity=0.4,
    color=colors   # <-- colors assigned here, not in px.scatter_3d()
))

fig.update_layout(
    scene=dict(
        xaxis=dict(title='R', range=[0, 255]),
        yaxis=dict(title='G', range=[0, 255]),
        zaxis=dict(title='B', range=[0, 255]),
    ),
    width=600, height=500,
    margin=dict(l=0, r=0, b=0, t=40)
)

fig.show()
````

## Indizierung — Zeilen, Spalten, Kanäle

Python-Slice-Syntax: `start:stop` — **stop ist nicht mehr in der Menge der gewählten Elemente enthalten. Wie Timecodes in EDLs!**
Ein einzelner `:` bedeutet 'alle Elemente'.

| MATLAB | Python | Bedeutung |
|---|---|---|
| `img(1:80,:,:)` | `img[:80,:,:]` | erste 80 Zeilen |
| `img(end-80:end,:,:)` | `img[-80:,:,:]` | letzte 80 Zeilen |
| `img(:,:,1)` | `img[:,:,0]` | Rotkanal (Index **0**!) |
| `img(:,:,2)` | `img[:,:,1]` | Grünkanal |
| `img(:,:,3)` | `img[:,:,2]` | Blaukanal |

## Für die Testbildaufnahme mal schnell einen Siemensstern als Vektorgrafik erstellen?

````python
# Generate values from 0 to 2*pi
circlePoints = np.linspace(0, 2 * np.pi, 36*2+1)
 
fig, ax = plt.subplots()
 
# Draw all segments of the Siemens star
for ii in range(0, len(circlePoints) - 1, 2):
    ax.fill([0, np.sin(circlePoints[ii]), np.sin(circlePoints[ii + 1]), 0],
            [0, np.cos(circlePoints[ii]), np.cos(circlePoints[ii + 1]), 0],
            color='black', edgecolor='none')
 
# Format and save
ax.set_aspect('equal')
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
plt.title('Mein erster Siemensstern...')
plt.savefig('Siemensstern.pdf')
plt.show()
````

## Aus Gruenden der Performance und Lesbarkeit immer vektorisierten Code schreiben!

Beispiel: **Kontrastfunktion**  $\\text{out} = (\\text{in} - 0.5) \\times \\text{contrast} + 0.5$

Zwei Implementierungen — identisches Ergebnis, sehr unterschiedliche Geschwindigkeit.


````python
# Zuerst von uint8 (0-255) nach float64 (0-1) konvertieren
img_float = img.astype(np.float64) / 255.0
print('dtype:', img_float.dtype, '  Wertebereich:', img_float.min(), '...', img_float.max())
````

````python
contrast = 1.1

# -- For-Schleife ----------------------------------------------------------
H, W, C = img_float.shape

t0 = time.perf_counter()   # MATLAB: tic

img_for_loop = np.empty_like(img_float)
for a in range(H):
    for b in range(W):
        for c in range(C):
            img_for_loop[a,b,c] = (img_float[a,b,c] - 0.5) * contrast + 0.5

t_loop = time.perf_counter() - t0   # MATLAB: toc
print(f'For-Schleife benoetigt viel Zeit: {t_loop*1000:.0f} ms')
````

````python
# -- Vektorisierter Code --------------------------------------------------
t0 = time.perf_counter()

img_vec = (img_float - 0.5) * contrast + 0.5

t_vec = time.perf_counter() - t0
print(f'Vektorisiert geht schneller: {t_vec*1000:.2f} ms')
print(f'Speedup: {t_loop/t_vec:.0f}x')
````

````python
# MATLAB: if isequal(imgForLoop, imgVec)
if np.allclose(img_for_loop, img_vec):
    print('Ergebnis ist aber gleich')
else:
    print('Wir haben ein Problem!')
````

````python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
axes[0].imshow(np.clip(img_float, 0, 1)); axes[0].set_title('Original');             axes[0].axis('off')
axes[1].imshow(np.clip(img_vec,   0, 1)); axes[1].set_title(f'Kontrast = {contrast}'); axes[1].axis('off')
plt.tight_layout()
plt.show()
````

## Zusammenfassung — MATLAB -> Python Spickzettel

Fuer Studierende die bis WS2026 MATLAB gelernt haben)

| MATLAB | Python / NumPy | Hinweis |
|---|---|---|
| `a = 3.14` | `a = 3.14` | identisch |
| `a` ohne `;` | `a` als letzter Ausdruck | identisch |
| `a = x;` (unterdrücken) | `a = x` | Zuweisung ist immer still |
| `[1, 2, 3]` | `np.array([1, 2, 3])` | |
| `[1; 2; 3]` | `np.array([[1],[2],[3]])` | |
| `b'` (transponieren) | `b.reshape(-1,1)` | |
| `m(1,3)` | `m[0,2]` | **0-indiziert!** |
| `m(end,end)` | `m[-1,-1]` | negativ = vom Ende |
| `m.*m` | `m * m` | elementweise (Standard!) |
| `m*m` | `m @ m` | Matrixmultiplikation |
| `m.^2` | `m ** 2` | |
| `eye(3)` | `np.eye(3)` | |
| `uint8(x)` | `np.clip(x,0,255).astype(np.uint8)` | erst clip! |
| `double(img)/255` | `img.astype(np.float64)/255` | |
| `imread('f.png')` | `plt.imread('f.png')` | |
| `imshow(img)` | `ax.imshow(img)` + `plt.show()` | |
| `imwrite(img,'f.jpg')` | `plt.imsave('f.jpg', img)` | |
| `size(img)` | `img.shape` | |
| `img(1:80,:,:)` | `img[:80,:,:]` | stop ausgeschlossen |
| `img(end-80:end,:,:)` | `img[-80:,:,:]` | |
| `img(:,:,1)` | `img[:,:,0]` | **0-indizierte Kanäle!** |
| `1:2:end` | `::2` | Schritt-Slice |
| `reshape(x,[],3)` | `x.reshape(-1,3)` | |
| `tic / toc` | `t=time.perf_counter()` / `time.perf_counter()-t` | |
| `isequal(a,b)` | `np.allclose(a,b)` | toleriert Float-Rundung |
